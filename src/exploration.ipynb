```python
#EXPLORAÇÃO DA BASE CUSTOMERS

import pandas as pd

from load import ler_parquet

df_customers = ler_parquet(bucket_path = "s3://lake-ecom-bronze-gclauar/customers")

display(df_customers.info())
display(df_customers.describe(include='all'))
display(df_customers.isna().sum())
display(df_customers.duplicated(subset=['customer_id']).sum())
display(df_customers['state'].value_counts())
display(df_customers['city'].value_counts())
```
